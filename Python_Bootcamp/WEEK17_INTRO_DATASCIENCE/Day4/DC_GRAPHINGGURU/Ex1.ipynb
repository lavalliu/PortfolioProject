```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(0,20, num=100, endpoint=True)
y = x * x

fig,ax=plt.subplots()
ax.set_title('Plot of x vs y (y=x^2)')
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,20)
ax.set_ylim(0,400)
plt.savefig('Task1.png')
```
