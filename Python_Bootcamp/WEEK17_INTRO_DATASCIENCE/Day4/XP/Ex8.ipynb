```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)  # Sine function
y2 = np.cos(x)  # Cosine function

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# Line plot on the first subplot
ax1.plot(x, y1, color='tab:blue', label='sin(x)')

# Scatter plot on the second subplot
ax2.scatter(x, y2, color='tab:green', marker='o', label='cos(x)')

plt.tight_layout()
plt.show()
```
