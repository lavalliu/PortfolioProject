```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

chipo=pd.read_csv('./chipotle.tsv', sep="\t")
print(chipo)
x = chipo['item_price'].apply(lambda x: float(x.replace('$','')))
y = chipo['quantity']

fig,ax=plt.subplots()
ax.set_title('Relationship between Item Price and Quantity Ordered')
ax.scatter(x,y)
ax.set_xlabel('Item Price')
ax.set_ylabel('Quantity Ordered')
plt.savefig('Task2.png')
```
