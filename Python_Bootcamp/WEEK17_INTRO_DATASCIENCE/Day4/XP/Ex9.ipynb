```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x) 

# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.plot(x, y1, color='tab:blue', label='y = sin(x)')
ax2.scatter(x, y2, color='tab:green', marker='o', label='y = cos(x)')
ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()
```
