```python
import random

def throw_dice():
    firstnum = secondnum = counter = counter2 = 0
    while counter2 < 100:
        firstnum = random.randint(1,6)
        secondnum = random.randint(1,6)
        counter += 1
        if firstnum == secondnum:
            print(f"First dice : {firstnum} - Second dice : {secondnum}")
            counter2 += 1
    print(f"The dice was thrown {counter} times")
    print(f"The average throws for each double is {round(counter/10, 2)}")

def throw_until_doubles():
    throw_dice()

def main():
    throw_until_doubles()
main()
```
