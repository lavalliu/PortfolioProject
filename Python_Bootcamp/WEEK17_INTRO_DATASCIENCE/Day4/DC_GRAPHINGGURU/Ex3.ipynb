```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

chipo=pd.read_csv("./chipotle.tsv", sep="\t")

totalbyitem = chipo.groupby('item_name')['quantity'].sum()
topitems = totalbyitem.sort_values(ascending=False).head(5)
x = topitems.index
y = topitems.values

fig,ax=plt.subplots()
ax.set_title('Top 5 items by quantity ordered')
ax.bar(x, y)
ax.set_xlabel('Item Name')
ax.set_ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')
plt.savefig('Test3.png', bbox_inches='tight')
```
