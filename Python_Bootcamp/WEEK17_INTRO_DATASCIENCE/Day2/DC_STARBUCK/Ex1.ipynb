```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

starbucks=pd.read_csv('./starbucks.csv', sep="\,")
starbucks.head(10)
```

```python
print(f"Number of rows {starbucks.shape[0]}")
```

```python
# if there is any index on the df
starbucks.index.names
```

```python
starbucks.sort_values(by=' Sugars (g)', ascending=False, inplace=True)
starbucks.head()
```

```python
average_sugars = starbucks.groupby('Beverage')[' Sugars (g)'].mean()

most_sugar_beverage = average_sugars.idxmax()
most_sugar_average = average_sugars.max()

print(f"The drink with the most amount of sugar on average is {most_sugar_beverage} with an average of {most_sugar_average} grams of sugar.")
```

```python
grouped_bev = starbucks.groupby('Beverage_category')
print(grouped_bev.first())
```

```python
average_calories = starbucks.groupby('Beverage_category')['Calories'].mean()

# Display the average calorie content per drink category
print(average_calories)
```

```python
starbucks['Caffeine (mg)']=starbucks['Caffeine (mg)'].fillna(0)
starbucks.info()
```

```python
# Replace 'Varies' with NaN
starbucks['Caffeine (mg)'] = starbucks['Caffeine (mg)'].replace('Varies', np.nan)

# Convert the column to numeric, coercing errors to NaN
starbucks['Caffeine (mg)'] = pd.to_numeric(starbucks['Caffeine (mg)'], errors='coerce')
```

```python
starbucks['caffeine_to_calories'] = starbucks['Caffeine (mg)'] / starbucks['Calories']
starbucks.head()
```

```python
starbucks = starbucks.drop(starbucks[(starbucks['Caffeine (mg)'] == 0)].index)
high_caf = starbucks.sort_values(by='caffeine_to_calories', ascending=False)

print(f"The beverage that, on average, has the highest ‘caffeine_to_calories’ ratio is {high_caf['Beverage'].iloc[0]}")
print(f"The beverage category that, on average, has the highest ‘caffeine_to_calories’ ratio is {high_caf['Beverage_category'].iloc[0]}")
```

```python
# Calculate the 'caffeine_to_calories' ratio
starbucks['caffeine_to_calories'] = starbucks['Caffeine (mg)'] / starbucks['Calories']

# Calculate the average ratio for each category
average_ratios = starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean()
average_ratios = average_ratios.sort_values()

# Create the horizontal bar plot
plt.barh(average_ratios.index, average_ratios.values)

plt.xlabel('Average Caffeine to Calories Ratio')
plt.ylabel('Drink Category')
plt.title('Average Caffeine to Calories Ratio by Drink Category')
plt.show()
```

```python
starbucks['caffeine_to_sugar'] = starbucks['Caffeine (mg)'] / starbucks[' Sugars (g)']
high_sugar = starbucks.sort_values(by='caffeine_to_sugar', ascending=True)
high_sugar.head()
print(f"The beverage that, on average, has the highest ‘caffeine_to_sugar’ ratio is {high_sugar['Beverage'].iloc[0]}")
print(f"The beverage category that, on average, has the highest ‘caffeine_to_calories’ ratio is {high_sugar['Beverage_category'].iloc[0]}")
```
